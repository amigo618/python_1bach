## Entradas y salidas de datos
En Python hay dos métodos básicos que permiten introducir y visualizar datos
* **input**: Permite al usuario aportar datos al programa en "tiempo real"
* **print**: Devuelve en pantalla la información especificada

```python
print("Hola mundo")
```

## Tipos de variables
Python permite trabajar con múltiples tipos de datos: textuales, numéricos, estructurados, etc. Pero esta información debe almacenarse en algún sitio (un espacio en memoria). Esto es lo que llamamos una "variable".

En función de lo que queramos representar, y del modo en el que queramos trabajar, tenemos multiples opciones:
  
Numéricas
* **int**: Números enteros - 1, 2, 3, etc.
* **bool**: Numeros binarios - Verdadero/Falso (1/0)
* **float**: Números decimales (aproximaciones binarias) - 1.618, 3.14, 2.71, etc.

Textuales
* **str**: Cadenas de caracteres (letras, palabras, frases, etc.) - "string"
  
Estructuradas
* **listas**: Lista de variables (pueden ser numéricos, textuales, etc.) - [val1, val2, val2, etc.]
* **diccionarios**: Lista de clave-valor (pueden ser numéricos, textuales, etc.) - {key1:val1, key2:val2, key3:val3, etc.}

```python
## Todo esto son 1s
print(f"Esto es un 1 ---> {int(1)}")
print(f"Esto también lo es ---> {float(1)}")
print(f"Y por supuesto esto también ---> {bool(1)}")
```

## Operaciones numéricas en ordenadores
### Operadores Básicas:
| Operador | Nombre |
| :--- | :--- | 
| `+` | Suma
| `-` | Resta
| `*` | Multiplicación
| `/` | División
| `**` | Potencia

### Operadores Especiales:
| Operador | Nombre | Función |
| :--- | :--- | :--- |
| `//` | División Entera | Quita los decimales del resultado. |
| `%` | Módulo | Devuelve el resto de una división. |
| `**` | Potencia | Eleva un número a otro. |

```python
print("Cargando calculadora...")
print("-----------------------")

a = int(input("Introduce un entero"))
b = int(input("Introduce otro entero"))
print("-----------------------")

print(f"Su suma es: {a+b}")
print(f"Su resta es: {a-b}")
print(f"Su multiplicación es: {a*b}")
print(f"Su división es: {a/b}")
print(f"Su potencia es: {a**b}")
```

```python
# El curioso caso de los decimales - Se almacenan en binario, como una aproximación de división de potencias de 2 (0.5, 0.25, 0.125, etc.)
print(f"\n¿Cuánto es 0.1 + 0.2 en Python? {0.1 + 0.2}")
```
