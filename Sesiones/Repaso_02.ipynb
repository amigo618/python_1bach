### Termostato inteligente (Máquina de estados)
La calefacción se enciende si hace frío (menos de 18º) y se apaga cuando ya hace calor (más de 22º). ¿En qué estado termina el sistema?

```python
estado_calefaccion = "apagada"
temperaturas_sala = [15, 19, 23, 20]

for t in temperaturas_sala:
    if t < 18 and estado_calefaccion == "apagada":
        estado_calefaccion = "encendida"
    elif t > 22 and estado_calefaccion == "encendida":
        estado_calefaccion = "apagada"

print(estado_calefaccion)
```

### Radar de aceleración
Un radar mide la velocidad de un coche cada segundo. Queremos contar cuántas veces el coche ha acelerado respecto a la medición anterior.

```python
velocidades = [50, 60, 60, 75, 70]
aceleraciones = 0
velocidad_anterior = 50

for v in velocidades:
    if v > velocidad_anterior:
        aceleraciones = aceleraciones + 1
    velocidad_anterior = v

print(aceleraciones)
```

### Control de pérdidas (Nivel de un embalse)
Simulamos la caída desde el punto más alto del nivel de agua de un embalse a lo largo del tiempo. ¿Cuál fue la máxima pérdida registrada respecto al pico máximo previo?

```python
niveles = [1000, 1500, 1200, 1600]
pico_maximo = 0
perdida_maxima = 0

for n in niveles:
    if n > pico_maximo:
        pico_maximo = n
    
    perdida_actual = pico_maximo - n
    
    if perdida_actual > perdida_maxima:
        perdida_maxima = perdida_actual

print(perdida_maxima)
```
