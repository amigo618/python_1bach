Aqu√≠ ten√©is un peque√±o resumen de las funciones que os ayudar√°n a solucionar los retos propuestos:

### Casting (transformaci√≥n de datos)

- **str (numero)**: Convierte un n√∫mero en texto.
- **int(texto) / float(texto)**: Convierte el texto en n√∫mero.

### Strings (cadenas de Texto)

- **texto[inicio:final]**: Devuelve un nuevo texto a partir del texto original, entre los √≠ndices indicados (a esto lo llamamos slicing).
- **texto.startswith("ejemplo")**: Devuelve "True" si el texto empieza por un prefijo espec√≠fico.
- **texto.isdigit()**: Devuelve "True" si el texto est√° compuesto solo por n√∫meros.
- **"valor" in texto**: Devuelve "True" si una palabra existe dentro del texto.
- **len(texto)**: Devuelve el n√∫mero total de caracteres del texto.

### Listas 

- **lista = []**: Crea una lista vac√≠a para acumular resultados.
- **lista.append(dato)**: A√±ade un elemento al final de la lista.
- **valor in lista**: Devuelve "True" si un dato espec√≠fico est√° en la lista.
- **len(lista)**: Devuelve el n√∫mero de elementos guardados.
- **sum(lista)**: Devuelve la suma todos los n√∫meros de la lista (s√≥lo si son todos n√∫meros).

### Bucles (for)

- **for elemento in lista**: Repite el c√≥digo por cada √≠tem de la lista.
- **break**: Detiene y sale del bucle inmediatamente (bot√≥n del p√°nico).
- **contador += 1**: Incrementa en 1 una variable externa (√∫til para errores).
- **Indentaci√≥n (Tabulador)**: Todo el c√≥digo dentro del bucle debe estar movido a la derecha.

### Condicionales (if)

- **if / elif / else**: Estructura para decidir entre uno o varios caminos.
- **== / !=**: Operadores para comparar si algo es igual o distinto.
- **> / < / >= / <=**: Operadores para comparar magnitudes num√©ricas.
**and**: Ambas condiciones deben cumplirse obligatoriamente.
**or**: Basta con que se cumpla una de las dos condiciones.
**not**: Invierte el resultado (ej: if not pieza.startswith("OK_")).

###¬†‚ö†Ô∏è Recordatorios Cr√≠ticos

- **Dos puntos (:)**: Obligatorios tras if, elif, else y for.
- **Comparaci√≥n**: Usa == para preguntar, nunca = (que es para guardar).
- **Transformaci√≥n**: Siempre usa int() antes de comparar un n√∫mero que viene de un input o un string.
