## 1. Teor√≠a Pr√°ctica I: Listas (El Orden Importa)
Las listas `[]` son como una fila de cajas numeradas. Sirven para guardar elementos donde el **orden** es lo m√°s importante.



### Ejemplo 1: Gesti√≥n de la cola de pedidos
Ejecuta esta celda para ver c√≥mo evoluciona una lista de pedidos.

```python
pedidos = ["Margarita", "Barbacoa", "Vegetal"]

pedidos.append("Hawaiana")          # Llega un nuevo pedido al final
primero = pedidos.pop(0)           # Sacamos el primer pedido para cocinarlo

print(f"Cocinando ahora: {primero}")
print(f"Pedidos pendientes (en orden): {pedidos}")
print(f"Siguiente en la cola: {pedidos[0]}")
```

## 2. Teor√≠a Pr√°ctica II: Diccionarios (El Significado Importa)
Los diccionarios `{}` no funcionan por n√∫mero de orden, sino por **Clave: Valor**. Son como una ficha t√©cnica de un producto.



### Ejemplo 2: La ficha t√©cnica de una Pizza
Aqu√≠ no nos importa el orden, sino acceder al dato exacto (precio, ingredientes...).

```python
pizza = {
    "nombre": "Especial Amigo618",
    "precio": 12.50,
    "disponible": True,
    "ingredientes": ["Bacon", "Huevo", "Queso"]
}

print(f"Nombre: {pizza['nombre']}")
print(f"¬øCu√°ntos ingredientes lleva?: {len(pizza['ingredientes'])}")

# Modificar un valor
pizza["precio"] = 13.50
print(f"Nuevo precio: {pizza['precio']}‚Ç¨")
```
