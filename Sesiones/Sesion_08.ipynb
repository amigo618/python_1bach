## 游깱 Parte 1: Composici칩n de Ondas

Partamos de la definici칩n de onda que trabajamos el otro d칤a. Pero, en este caso, vamos a incluirle arm칩nicos de mayor frecuencia y "ruido".
La ecuaci칩n de la onda base con arm칩nicos es: 
$$y = \sin(x) + \frac{1}{3}\sin(3x) + \frac{1}{5}\sin(5x)$$
La ecuaci칩n de la onda con ruidos, en cambio, es la onda base con un componente aleatorio a침adido: 
$$y = \sin(x) + ruido$$

### 1.1 El Gr치fico Base (Sin Est칠tica)

```python
# Primero importamos las librerias que vamos a usar
import numpy as np 
import matplotlib.pyplot as plt


def pintamelon(funcion1, funcion2):
    # Gr치fico funcional pero aburrido
    plt.plot(x, funcion1)
    plt.plot(x, funcion2)
    plt.show()


tipo = "ruido" #마rmonico / ruido

x = np.linspace(0, 2*np.pi, 500) # Creo el vector de valores del eje X (500 puntos entre 0 y 2**np.pi)
y1 = np.sin(x) # Ahora aplico la funci칩n base

if tipo == "ruido":
    y2 = np.random.normal(0, 0.25, 500) #ㅁplicamos ruido (media 0, desviaci칩n 0.25, 500 puntos)
    y_sum = y1 + y2
else:
    y2 = (1/3)*np.sin(3*x) # Arm칩nico de 3er orden
    y3 = (1/5)*np.sin(5*x) # Arm칩nico de 5췈 orden
    y_sum = y1 + y2 + y3

pintamelon(y1, y_sum)
```

### 1.2 El "Glow Up" Matem치tico
Ahora aplicamos: 
* **Colores Hexadecimales** para un look moderno.
* **LaTeX** en los t칤tulos.
* **Grosor de l칤nea (lw)** y **Transparencia (alpha)** para crear jerarqu칤a visual.

```python
plt.figure(figsize=(10, 5))

# Onda base (m치s fina y transparente)
plt.plot(x, y1, label=r"$\sin(x)$", color="#95a5a6", linestyle="--", lw=1.5, alpha=0.6)

# Onda resultante (fuerte y clara)
plt.plot(x, y_sum, label="Se침al Resultante", color="#2e86c1", lw=3)

# Detalles de calidad
plt.title(r"Superposici칩n de Ondas: $y = \sin(x) + \frac{1}{3}\sin(3x) + \frac{1}{5}\sin(5x)$", fontsize=14)
plt.xlabel("Fase (rad)")
plt.ylabel("Amplitud")
plt.grid(color='lightgray', linestyle=':', linewidth=0.5)
plt.legend(frameon=True, shadow=True)
plt.tight_layout()
plt.show()
```

## 游늵 Parte 2: Estad칤stica y Control de Calidad

El histograma nos permite ver si un proceso industrial est치 centrado o si hay demasiada dispersi칩n.

### 2.1 El Histograma Profesional
Analizamos 1000 piezas que deber칤an medir **50 mm**.

```python
# Generamos datos realistas (Media 50, Desviaci칩n 0.8)
medidas = np.random.normal(50, 0.8, 1000)

# Cambiamos el estilo global a uno m치s t칠cnico
plt.style.use('bmh')
plt.figure(figsize=(10, 5))

# Histograma con bordes blancos para separar barras
plt.hist(medidas, bins=35, color='#27ae60', edgecolor='white', alpha=0.7)

# L칤neas de Ingenier칤a (Tolerancias)
plt.axvline(50, color='darkred', lw=2, label="Valor Nominal (50mm)")
plt.axvline(48.5, color='orange', ls='--', lw=1.5, label="L칤mite Inferior")
plt.axvline(51.5, color='orange', ls='--', lw=1.5, label="L칤mite Superior")

plt.title("An치lisis de Distribuci칩n: Control de Calidad", fontsize=15)
plt.xlabel("Medida (mm)")
plt.ylabel("Cantidad de Piezas")
plt.legend()
plt.show()
```
