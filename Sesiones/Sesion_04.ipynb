## 1. ¬øQu√© es realmente un condicional?

Es una estructura que eval√∫a una **expresi√≥n booleana** (algo que solo puede ser Verdadero o Falso). 



* **`if` (Si...):** La condici√≥n principal.
* **`elif` (Si no, pero si...):** Condiciones alternativas si la anterior fall√≥.
* **`else` (En cualquier otro caso):** Lo que ocurre si nada de lo anterior se cumpli√≥.

Imagina un laboratorio que etiqueta muestras peligrosas con una 'X'.
