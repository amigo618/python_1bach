## 1. Concepto de "Caja Negra" y √Åmbito (Scope)

Una funci√≥n es un compartimento estanco. Lo que pasa dentro de la funci√≥n, se queda dentro. 

* **Variables Locales:** Una variable creada dentro de una funci√≥n solo existe all√≠. Al terminar la funci√≥n, desaparece. 
* **Independencia:** Tu programa principal no necesita saber *c√≥mo* la funci√≥n calcula el resultado, solo qu√© datos enviarle y qu√© esperar de vuelta.



--- 
## 2. Par√°metros por Defecto

Podemos configurar una funci√≥n para que tenga valores est√°ndar si el usuario no especifica uno. Esto hace que nuestro c√≥digo sea m√°s flexible y robusto.

```python
def saludar_estacion(nombre, planeta="Marte"):
    print(f"Hola {nombre}, bienvenido a la base de {planeta}")

saludar_estacion("Nacho")            # Usar√° "Marte" por defecto
saludar_estacion("Elena", "Saturno") # Usar√° "Saturno"
```

## 3. La diferencia cr√≠tica: `print()` vs `return` 

* **`print()`**: Solo muestra un mensaje en pantalla para el humano. El ordenador **no guarda** ese dato.
* **`return`**: Env√≠a el dato de vuelta al programa principal. Es el "producto" que sale de la f√°brica y que podemos guardar en una variable.
