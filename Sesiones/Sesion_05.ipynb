## 1. El Concepto de Iteraci√≥n
Imagina una cinta transportadora en una f√°brica. Cada objeto que pasa es un elemento de una **Lista**. El bucle `for` es el operario que agarra cada objeto, hace algo con √©l y lo suelta para agarrar el siguiente.



### Ejemplo 1: El Recorrido Simple
Observa c√≥mo la variable `planeta` va tomando el valor de cada eslab√≥n de la cadena autom√°ticamente.

```python
sistema_solar = ["Mercurio", "Venus", "Tierra", "Marte"]

for planeta in sistema_solar:
    print(f"Explorando el planeta: {planeta}")
    
print("üõ∞Ô∏è Misi√≥n de reconocimiento finalizada.")
```

## 2. La funci√≥n `range()`: El Contador de Pulsos
A veces no queremos recorrer una lista, sino simplemente repetir una acci√≥n un n√∫mero exacto de veces (como hacer parpadear un LED en **Arduino**).

La funci√≥n `range(inicio, fin)` genera una secuencia de n√∫meros. 
*Nota: El n√∫mero final nunca se incluye.*

```python
print("Iniciando cuenta atr√°s para el lanzamiento:")

for i in range(10, 0, -1):  # Empieza en 10, acaba en 1, baja de 1 en 1
    print(f"üöÄ T-minus {i}")

print("¬°IGNICI√ìN!")
```

## 3. Acumuladores: El Bucle con Memoria
Uno de los usos m√°s potentes del `for` es ir guardando resultados de vueltas anteriores. Para esto usamos una variable **acumuladora** fuera del bucle.

### Ejemplo 3: Sumatorio de Telemetr√≠a
Vamos a calcular el consumo total de combustible de una nave.

```python
consumos_motores = [120, 150, 110, 130] # Litros por segundo
total_consumido = 0

for motor_gasto in consumos_motores:
    total_consumido = total_consumido + motor_gasto
    print(f"Sumando motor... Total provisional: {total_consumido}")

print(f"--- Consumo total final: {total_consumido} L/s ---")
```

## 4. Bloque de Debugging üõ†Ô∏è
**Misi√≥n:** Un ingeniero quiere limpiar una lista de lecturas de sensores que tienen espacios vac√≠os y errores de min√∫sculas, pero el c√≥digo no hace nada. ¬øPor qu√©?

```python
sensores = [" temp_01", "hum_02 ", "PRES_03"]

for s in sensores:
    s.strip()
    s.upper()

print("Lista de sensores corregida:")
print(sensores) # Deber√≠a salir ['TEMP_01', 'HUM_02', 'PRES_03']
```
