# La Iteraci√≥n: El Bucle `for`
--- 

Los bucles son una de las herramientas m√°s potentes en programaci√≥n. Nos permiten ejecutar un bloque de c√≥digo **muchas veces** de forma autom√°tica.
Hasta ahora, si quer√≠as procesar 10 datos, ten√≠as que escribir 10 l√≠neas de c√≥digo. Con bucles, esto cambia completamente.


## El Concepto de Iteraci√≥n
Imagina un brazo rob√≥tico en una cinta transportadora en una f√°brica, d√≥nde cada objeto que pasa es un elemento de una **Lista**. El brazo lo coge, lleva a cabo diferentes operaciones con √©l, y pasa al siguiente.


Observa c√≥mo la variable `planeta` va tomando el valor de cada eslab√≥n de la cadena autom√°ticamente.

```python
sistema_solar = ["Mercurio", "Venus", "Tierra", "Marte"]

for planeta in sistema_solar:
    print(f"Explorando el planeta: {planeta}")
    
print("Misi√≥n de reconocimiento finalizada.")
```

## La funci√≥n `range()`
A veces no queremos recorrer una lista, sino simplemente repetir una acci√≥n un n√∫mero exacto de veces (como hacer parpadear un LED en **Arduino**).

La funci√≥n `range(inicio, fin)` genera una secuencia de n√∫meros. Podemos trabajar con esa secuencia, utilizarla como √≠ndice, o simplemente guardarla.

```python
cuenta = [] # Generamos una lista vac√≠a

for i in range(5,10):
    cuenta.append(i) # Metemos cada terminoo en la lista, uno a uno
    
print(cuenta)
```

Adem√°s de especificar el rango o intervalo, podemos hacer que se recorra en sentido inverso. Para eso, debemos de incluir un paso: `range(inicio, fin, paso)`

```python
print("Iniciando cuenta atr√°s para el lanzamiento:")

for i in range(10, 0, -1):  # Empieza en 10, acaba en 1, baja de 1 en 1
    print(f"T-minus {i}")

print("¬°IGNICI√ìN!")
```

## Acumuladores: Bucles con Memoria
Uno de los usos m√°s potentes del `for` es ir guardando resultados de vueltas anteriores. Para esto usamos una variable **acumuladora** fuera del bucle.

Vamos a calcular el consumo total de combustible de una nave.

```python
consumos_motores = [120, 150, 110, 130] # Litros por segundo
total_consumido = 0

for motor_gasto in consumos_motores:
    total_consumido = total_consumido + motor_gasto
    print(f"Sumando motor... Total provisional: {total_consumido}")

print(f"--- Consumo total final: {total_consumido} L/s ---")
```
