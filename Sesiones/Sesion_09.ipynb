## Finanzas en Tiempo Real (`yfinance`)

Yahoo Finance tiene una de las APIs m√°s potentes para ingenieros. Vamos a bajar el hist√≥rico de una criptomoneda o acci√≥n y visualizarlo.

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Pedir datos al servidor
activo = "BTC-USD" # Prueba con "TSLA" o "NVDA"
datos = yf.download(activo, start="2024-01-01", end="2026-01-01")
print(datos.columns)

# Gr√°fico Pro
plt.style.use('ggplot')
plt.figure(figsize=(12, 5))
plt.plot(datos['High'], color='#2ecc71', lw=2, label="Precio Cierre")
plt.title(f"Evoluci√≥n Hist√≥rica de {activo}", fontsize=16)
plt.grid(ls=':')
plt.legend()
plt.show()
```

## Meteorolog√≠a y JSON (`Open-Meteo`)

Muchas APIs no tienen una librer√≠a propia y hay que hablar con ellas mediante "peticiones web" (`requests`). Los datos suelen venir en formato **JSON** (un formato parecido a un diccionario de Python, y con el que podemos trabajar como si lo fuese, con los mismos m√©todos y l√≥gica de los diccionarios).

```python
import requests

# URL de la API (Coordenadas de Madrid)
url = "https://api.open-meteo.com/v1/forecast?latitude=40.41&longitude=-3.70&hourly=temperature_2m"

# Hacemos la petici√≥n
respuesta = requests.get(url)
datos_clima = respuesta.json()

# Extraemos la lista de temperaturas (navegando por el diccionario)
temps = datos_clima['hourly']['temperature_2m']
horas = range(len(temps))

# Visualizaci√≥n
plt.figure(figsize=(10, 4))
plt.plot(horas, temps, color='orange', marker='.')
plt.title("Previsi√≥n de Temperatura (Pr√≥ximas 168h)")
plt.xlabel("Horas desde ahora")
plt.ylabel("¬∫C")
plt.show()
```

## L√≥gica y Estructura (`PokeAPI`)

Hay decebas de APIs ah√≠ fuera, es la opci√≥n preferida por empresas y serivicios para comunicaciones autom√°ticas. Por fortuna, tambi√©n hay varias que nos sirven solo "para practicar", y as√≠ aprender a trabajar y filtar con **bucles y condicionales** sobre datos reales y complejos.

```python
nombre = "pidgey"
url_poke = f"https://pokeapi.co/api/v2/pokemon/{nombre}"
data = requests.get(url_poke).json()

print(data)

print(f"--- Reporte de Estad√≠sticas: {nombre.upper()} ---")

# L√ìGICA: Recorremos la lista de estad√≠sticas y aplicamos un filtro
for s in data['stats']:
    stat_nombre = s['stat']['name']
    valor = s['base_stat']
    
    # Condicional de estado
    if valor > 90:
        estado = "üî• ¬°NIVEL √âPICO!"
    elif valor > 50:
        estado = "‚úÖ Estable"
    else:
        estado = "‚ö†Ô∏è Debilidad detectada"
        
    print(f"{stat_nombre.capitalize()}: {valor} -> {estado}")
```
